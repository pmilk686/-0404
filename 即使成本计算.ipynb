## 引入每日即时成本文件

```python
fresh_data = pd.read_csv("C:\\Users\\86130\\Desktop\\2024_0401_meituan.csv",encoding='ANSI')
```

```python
fresh_data.head()
```

```python
fresh_data = fresh_data.iloc[:,[1,5]]
```

#### 把菜品名称设置为索引

```python
fresh_data = fresh_data.rename(columns={"商品名" : "菜品名称"})
fresh_data = fresh_data.set_index("菜品名称")
fresh_data = fresh_data.sort_index()
```

```python
final_data = pd.merge(new_data,fresh_data,on = "菜品名称",how = "outer")
```

##### 计算菜品总成本

```python
final_data["菜品总成本"] =  final_data["单个成本/元"]*final_data["商品销量"]
final_data = final_data.sort_index()
```

#### 合并重复数据

```python
final_data = final_data.groupby('菜品名称').mean()
```

# 最终导出表格

```python
final_data.to_csv("2024_0401_meituan.csv",index=True)
```
